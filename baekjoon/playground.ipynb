```python
# 11723 "집합" - 비트 연산
import sys

input = sys.stdin.readline

m = int(input())
mask = 0
out = []

for _ in range(m):
    line = input().strip()
    if line == "all":
        mask = (1 << 21) - 2   # 비트 1..20 ON (0번 비트는 미사용)
    elif line == "empty":
        mask = 0
    else:
        op, xs = line.split()
        x = int(xs)
        bit = 1 << x
        if op == "add":
            mask |= bit
        elif op == "remove":
            mask &= ~bit
        elif op == "check":
            out.append('1' if (mask & bit) else '0')
        elif op == "toggle":
            mask ^= bit

sys.stdout.write("\n".join(out))
```

```python
# 11723 "집합" - set 사용
import sys

input = sys.stdin.readline

m = int(input())
s = set()
out = []

for _ in range(m):
    line = input().strip()
    if line == "all":
        s = set(range(1, 21))
    elif line == "empty":
        s.clear()
    else:
        op, xs = line.split()
        x = int(xs)
        if op == "add":
            s.add(x)
        elif op == "remove":
            s.discard(x)
        elif op == "check":
            out.append('1' if x in s else '0')
        elif op == "toggle":
            if x in s:
                s.remove(x)
            else:
                s.add(x)

sys.stdout.write("\n".join(out))
```

```python
# 2108 "통계학 성공"

import sys

def solve():
    # 1. 입력 처리: N과 N개의 정수들을 입력받습니다.
    n = int(sys.stdin.readline())
    numbers = []
    for _ in range(n):
        numbers.append(int(sys.stdin.readline()))

    # 2. 산술 평균 계산
    # round() 함수는 0.5일 때 짝수로 반올림하는 특징(사사오입)이 있습니다.
    # 따라서, 올바른 반올림을 위해 0.5를 더하거나 빼는 트릭을 사용합니다.
    avg = sum(numbers) / n
    if avg >= 0:
        print(int(avg + 0.5))
    else:
        print(int(avg - 0.5))

    # 3. 중앙값 계산
    # 중앙값을 찾기 위해 리스트를 정렬합니다.
    numbers.sort()
    median = numbers[n // 2]
    print(median)

    # 4. 최빈값 계산
    # 각 숫자의 빈도수를 저장하기 위한 딕셔너리(frequency)를 사용합니다.
    # 입력 정수의 절댓값이 4000을 넘지 않으므로, -4000부터 4000까지의 인덱스를 갖는 배열로도 가능합니다.
    # 여기서는 딕셔너리를 사용하여 메모리를 효율적으로 사용합니다.
    frequency = {}
    for num in numbers:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1

    # 최빈 빈도수를 찾습니다.
    max_freq = 0
    for freq in frequency.values():
        if freq > max_freq:
            max_freq = freq

    # 최빈값을 모두 찾아서 리스트에 저장합니다.
    modes = []
    for num, freq in frequency.items():
        if freq == max_freq:
            modes.append(num)
    
    modes.sort()

    # 최빈값이 여러 개일 경우, 두 번째로 작은 값을 출력합니다.
    if len(modes) > 1:
        print(modes[1])
    else:
        print(modes[0])

    # 5. 범위 계산
    # 정렬된 리스트의 가장 큰 값과 가장 작은 값의 차이를 구합니다.
    _range = numbers[-1] - numbers[0]
    print(_range)

solve()
```

```python
# 1874 "스택 수열"

import sys
input = sys.stdin.readline

n = int(input())
targets = [int(input()) for _ in range(n)]

stack = []
ops = []
cur = 1          # 다음에 push할 수
possible = True

for t in targets:
    # t가 스택 top에 올 수 있게 t까지 push
    while cur <= t:
        stack.append(cur)
        ops.append('+')
        cur += 1
    # 이제 top이 t여야 pop 가능
    if stack and stack[-1] == t:
        stack.pop()
        ops.append('-')
    else:
        possible = False
        break

print('\n'.join(ops) if possible else 'NO')
```

```python
# 7576 "토마토"

import sys
from collections import deque
input = sys.stdin.readline

M, N = map(int, input().split())      # M: columns, N: rows
box = [list(map(int, input().split())) for _ in range(N)]

q = deque()
# 4-direction (row, col)
dr = [-1, 1, 0, 0]
dc = [0, 0, -1, 1]

# 초기: 익은 토마토 전부 큐에 넣기
for r in range(N):
    for c in range(M):
        if box[r][c] == 1:
            q.append((r, c))

# BFS
while q:
    r, c = q.popleft()
    for k in range(4):
        nr, nc = r + dr[k], c + dc[k]
        if 0 <= nr < N and 0 <= nc < M and box[nr][nc] == 0:
            box[nr][nc] = box[r][c] + 1   # 날짜 누적
            q.append((nr, nc))

# 결과 계산
ans = 0
for r in range(N):
    for c in range(M):
        if box[r][c] == 0:          # 익지 못한 토마토 존재
            print(-1)
            sys.exit(0)
        ans = max(ans, box[r][c])

print(ans - 1)   # 처음 익은 토마토가 1이므로 -1
```

```python
# 1260 "DFS와 BFS"

import sys
from collections import deque
input = sys.stdin.readline

# 입력은 N M V (정점 수, 간선 수, 시작 정점)
N, M, V = map(int, input().split())

adj = [[] for _ in range(N + 1)]
for _ in range(M):
    a, b = map(int, input().split())
    adj[a].append(b)
    adj[b].append(a)

# 정점 번호가 작은 것부터 방문해야 하므로 정렬
for i in range(1, N + 1):
    adj[i].sort()

# -------- DFS (스택) --------
dfs_order = []
visited = [False] * (N + 1)
stack = [V]

while stack:
    node = stack.pop()
    if visited[node]:
        continue
    visited[node] = True
    dfs_order.append(node)
    # 스택은 LIFO라, 작은 번호부터 꺼내려면 큰 번호부터 push
    for nxt in reversed(adj[node]):
        if not visited[nxt]:
            stack.append(nxt)

print(*dfs_order)

# -------- BFS (큐) --------
bfs_order = []
visited = [False] * (N + 1)
q = deque([V])
visited[V] = True

while q:
    node = q.popleft()
    bfs_order.append(node)
    for nxt in adj[node]:  # 오름차순 그대로
        if not visited[nxt]:
            visited[nxt] = True  # enqueue 시 방문표시
            q.append(nxt)

print(*bfs_order)
```

```python
# 1197 "최소 스패닝 트리"

import sys
input = sys.stdin.readline

v, e = map(int, input().split())
parent = [i for i in range(v+1)]
rank = [0] * (v+1)
edges = []
total = cnt = 0

for _ in range(e):
    a, b, c = map(int, input().split())
    edges.append((c, a, b))

edges.sort()

def find(x):
    if parent[x] != x:
        parent[x] = find(parent[x])
    return parent[x]

def union(x, y):
    x = find(x)
    y = find(y)
    if x == y:
        return False
    if rank[x] > rank[y]:
        parent[y] = x
        rank[x] += 1
    else:
        parent[x] = y
        rank[y] += 1
    return True

for c, a, b in edges:
    if union(a, b):
        total += c
        cnt += 1
        if cnt == v-1:
            print(total)
            break
```

```python
# 7662 "이중 우선순위 큐"

from heapq import heappush, heappop
import sys
input = sys.stdin.readline

T = int(input())
for _ in range(T):
    k = int(input())
    min_h, max_h = [], []
    alive = [False] * k   # 각 삽입 id의 생존 여부
    idx = 0               # 다음 삽입 id

    for _ in range(k):
        op, x = input().split()
        x = int(x)

        if op == 'I':
            heappush(min_h, (x, idx))
            heappush(max_h, (-x, idx))
            alive[idx] = True
            idx += 1

        else:  # 'D'
            if x == 1:
                # 최댓값 삭제
                while max_h and not alive[max_h[0][1]]:
                    heappop(max_h)
                if max_h:
                    _, i = heappop(max_h)
                    alive[i] = False
            else:
                # 최솟값 삭제
                while min_h and not alive[min_h[0][1]]:
                    heappop(min_h)
                if min_h:
                    _, i = heappop(min_h)
                    alive[i] = False

    # 최종 정리: 두 힙의 stale 노드 제거
    while min_h and not alive[min_h[0][1]]:
        heappop(min_h)
    while max_h and not alive[max_h[0][1]]:
        heappop(max_h)

    if not min_h or not max_h:
        print("EMPTY")
    else:
        print(-max_h[0][0], min_h[0][0])
```

```python
# 1978 "소수 찾기"

_ = input()
l = set(map(int, input().split()))

def eratos_sieve(n):
    # n+1 크기의 boolean 리스트를 생성하고, 모든 값을 True로 초기화
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False  # 0과 1은 소수가 아님

    # 2부터 n의 제곱근까지 반복
    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]:
            # i의 배수들을 모두 False로 변경
            for j in range(i * i, n + 1, i):
                is_prime[j] = False
    
    # 소수 리스트 생성
    primes = [i for i, prime in enumerate(is_prime) if prime]
    return primes

# 1000 이하의 소수 찾기
primes_under_1000 = set(eratos_sieve(1000))
print(len(l & primes_under_1000))
```

```python
# 1406 "에디터"

from collections import deque
import sys
input = sys.stdin.readline

left = deque(input().rstrip())
right = deque()
n = int(input())

for _ in range(n):
    cmd = input().rstrip()
    if cmd[0] == 'P':
        left.append(cmd[2])
    elif cmd[0] == 'L':
        if left:
            right.appendleft(left.pop())
    elif cmd[0] == 'D':
        if right:
            left.append(right.popleft())
    elif cmd[0] == 'B':
        if left:
            left.pop()

print(''.join(left) + ''.join(right))
```

```python
# 1759 "암호 만들기"

L, C = map(int, input().split())
letters = sorted(input().split())

vowels = {'a', 'e', 'i', 'o', 'u'}
path = []   # 현재 만들고 있는 암호 조합

def dfs(start):
    # 1) 길이 L이면 조건 체크 후 출력
    if len(path) == L:
        vowel_count = sum(1 for c in path if c in vowels)
        consonant_count = L - vowel_count
        if vowel_count >= 1 and consonant_count >= 2:
            print(''.join(path))
        return

    # 2) start부터 C까지 반복하며 DFS 확장
    for i in range(start, C):
        path.append(letters[i])   # 선택
        dfs(i + 1)                # 다음 문자부터 고르기
        path.pop()                # 선택 취소(백트래킹)

# DFS 시작
dfs(0)
```

```python
# 13023 "ABCDE"

import sys
sys.setrecursionlimit(10**7)
input = sys.stdin.readline

N, M = map(int, input().split())
graph = [[] for _ in range(N)]

# 친구 관계 입력
for _ in range(M):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

visited = [False] * N
answer = 0

def dfs(node, depth):
    global answer
    if depth == 4:   # A-B-C-D-E 연결 (depth 4면 노드 5개)
        answer = 1
        return

    visited[node] = True

    for nxt in graph[node]:
        if not visited[nxt]:
            dfs(nxt, depth + 1)
            if answer:     # 이미 답을 찾았으면 종료
                return

    visited[node] = False   # 백트래킹

# 모든 노드를 시작점으로 DFS 시도
for i in range(N):
    dfs(i, 0)
    if answer:
        break

print(answer)
```

```python
# 17618 "신기한 수"

def digit_sum(x):
    s = 0
    while x:
        s += x % 10
        x //= 10
    return s

n = int(input())
cnt = 0

digit_sum = 0

for i in range(1, n + 1):
    x = i
    digit_sum += 1

    while x % 10 == 0:
        digit_sum -= 9
        x //= 10

    if i % digit_sum == 0:
        cnt += 1

print(cnt)
```

```python
# 2057 "팩토리얼 분해"

import sys

def solve():
    line = sys.stdin.readline().strip()
    if not line:
        return
    n = int(line)

    # 예외 처리: 0은 팩토리얼의 합으로 만들 수 없음 (최소 0! = 1)
    if n == 0:
        print("NO")
        return

    # 1. 팩토리얼 리스트 미리 만들기 (20!까지)
    fact = [1] * 21
    for i in range(1, 21):
        fact[i] = fact[i-1] * i

    # 2. 그리디 알고리즘: 큰 값부터 빼기
    # i=20부터 0까지 역순으로 조사
    for i in range(20, -1, -1):
        if n >= fact[i]:
            n -= fact[i]

    # 3. 정확히 0이 되었는지 확인
    if n == 0:
        print("YES")
    else:
        print("NO")

if __name__ == "__main__":
    solve()
```

```python
# 9506 "약수들의 합"

while True:
    n = int(input())
    if n == -1:
        break

    divs = [1]
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            divs.append(i)
            if i != n // i:
                divs.append(n // i)

    divs.sort()

    if sum(divs) == n:
        print(f"{n} = " + " + ".join(map(str, divs)))
    else:
        print(f"{n} is NOT perfect.")
```

```python
# 11653 "소인수분해"

n = int(input())
divs = []

def prime_div(x):
    for i in range(2, int(x ** 0.5) + 1):
        if x % i == 0:
            return i
    return x

while n > 1:
    p = prime_div(n)
    divs.append(str(p))
    n //= p

print("\n".join(divs))
```

```python
# 11653 "소인수분해"

n = int(input())
d = 2
out = []

while d * d <= n:
    while n % d == 0:
        out.append(str(d))
        n //= d
    d += 1

if n > 1:
    out.append(str(n))

print("\n".join(out))
```

```python
# 28455 "Union Maplestory"

n = int(input())
levels = [int(input()) for _ in range(n)]

# 레벨 내림차순 정렬
levels.sort(reverse=True)

# 공격대 인원 (최대 42명)
team = levels[:42]

level_sum = sum(team)

def stat(level):
    cnt = 0
    if level >= 60: cnt += 1
    if level >= 100: cnt += 1
    if level >= 140: cnt += 1
    if level >= 200: cnt += 1
    if level >= 250: cnt += 1
    return cnt

stat_sum = sum(stat(lv) for lv in team)

print(level_sum, stat_sum)
```

```python
# 18258 "큐 2"

from collections import deque
import sys
input = sys.stdin.readline

n = int(input())
q = deque()
out = []

for _ in range(n):
    cmd = input().split()

    if cmd[0] == 'push':
        q.append(int(cmd[1]))
    elif cmd[0] == 'pop':
        out.append(str(q.popleft() if q else -1))
    elif cmd[0] == 'size':
        out.append(str(len(q)))
    elif cmd[0] == 'empty':
        out.append('0' if q else '1')
    elif cmd[0] == 'front':
        out.append(str(q[0] if q else -1))
    elif cmd[0] == 'back':
        out.append(str(q[-1] if q else -1))

sys.stdout.write("\n".join(out))
```

```python
# 28279 "덱 2"

from collections import deque
import sys
input = sys.stdin.readline

n = int(input())
d = deque()
out = []

for _ in range(n):
    cmd = input().split()

    if cmd[0] == '1':
        d.appendleft(int(cmd[1]))
    elif cmd[0] == '2':
        d.append(int(cmd[1]))
    elif cmd[0] == '3':
        out.append(str(d.popleft() if d else -1))
    elif cmd[0] == '4':
        out.append(str(d.pop() if d else -1))
    elif cmd[0] == '5':
        out.append(str(len(d)))
    elif cmd[0] == '6':
        out.append('0' if d else '1')
    elif cmd[0] == '7':
        out.append(str(d[0] if d else -1))
    elif cmd[0] == '8':
        out.append(str(d[-1] if d else -1))
        
sys.stdout.write("\n".join(out))
```

```python
# 9070 "장보기"

import sys
input = sys.stdin.readline

T = int(input())
out = []

for _ in range(T):
    N = int(input())
    best_w, best_c = 0, 1  # best_w/best_c 가 현재 최고 비율

    for _ in range(N):
        W, C = map(int, input().split())

        # W/C vs best_w/best_c 비교: W*best_c ? best_w*C
        left = W * best_c
        right = best_w * C

        if left > right or (left == right and C < best_c):
            best_w, best_c = W, C

    out.append(str(best_c))

print("\n".join(out))
```

```python
# 1049 "기타줄"

import sys

def main():
    input = sys.stdin.readline
    N, M = map(int, input().split())

    min_pack = 10**9
    min_single = 10**9

    for _ in range(M):
        pack, single = map(int, input().split())
        if pack < min_pack:
            min_pack = pack
        if single < min_single:
            min_single = single

    cost_all_single = N * min_single
    cost_all_pack = ((N + 5) // 6) * min_pack
    cost_mix = (N // 6) * min_pack + (N % 6) * min_single

    print(min(cost_all_single, cost_all_pack, cost_mix))

if __name__ == "__main__":
    main()
```

```python
# 14471 "포인트 카드"

import sys

def main():
    input = sys.stdin.readline
    N, M = map(int, input().split())

    costs = []
    for _ in range(M):
        A, B = map(int, input().split())
        costs.append(max(0, N - A))

    costs.sort()
    # M-1개 이상 경품 교환 가능해야 하므로, 가장 싼 (M-1)개를 만든다
    print(sum(costs[:M-1]))

if __name__ == "__main__":
    main()
```

```python
# 34115 "먼 카드"

import sys

def main():
    input = sys.stdin.readline
    N = int(input().strip())
    X = list(map(int, input().split()))

    first = [-1] * (N + 1)
    ans = 0

    for i, v in enumerate(X): 
        if first[v] == -1:
            first[v] = i
        else:
            ans = max(ans, i - first[v] - 1)

    print(ans)

if __name__ == "__main__":
    main()
```

```python
import sys

def main():
    input = sys.stdin.readline
    N = int(input().strip())

    sr = sc = -1
    legion = []

    for r in range(N):
        row = list(map(int, input().split()))
        for c, v in enumerate(row):
            if v == 2:
                sr, sc = r, c
            elif v == 1:
                legion.append((r, c))

    # 레기온이 없으면 바로 가능
    if not legion:
        print("Lena")
        return

    start_parity = (sr + sc) & 1
    for r, c in legion:
        if ((r + c) & 1) == start_parity:
            print("Kiriya")
            return

    print("Lena")

if __name__ == "__main__":
    main()
```

```python
# 1213 "팰린드롬 만들기"
import sys
from collections import Counter

def main():
    s = sys.stdin.readline().strip()
    cnt = Counter(s)

    # 홀수 개수 문자가 2개 이상이면 팰린드롬 불가
    odd_chars = [ch for ch, v in cnt.items() if v % 2 == 1]
    if len(odd_chars) > 1:
        print("I'm Sorry Hansoo")
        return

    mid = odd_chars[0] if odd_chars else ""

    # 사전순 최소를 만들기 위해 왼쪽 절반을 A~Z 순으로 구성
    left = []
    for ch in sorted(cnt.keys()):
        left.append(ch * (cnt[ch] // 2))
    left = "".join(left)

    print(left + mid + left[::-1])

if __name__ == "__main__":
    main()
```

```python
# 16676 "근우의 다이어리 꾸미기"
import sys

def main():
    N = int(sys.stdin.readline().strip())

    if N == 0:
        print(1)
        return

    k = 0
    rep = 0  # 1, 11, 111, ...
    while True:
        nxt = rep * 10 + 1
        if nxt <= N:
            rep = nxt
            k += 1
        else:
            break

    print(k)

if __name__ == "__main__":
    main()
```
